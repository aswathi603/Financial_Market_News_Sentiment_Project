```python
#1.importing library
import pandas as pd
import numpy as np
```

```python
#2. Loading the Dataset
df = pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Financial%20Market%20News.csv", encoding = "ISO-8859-1")
df.head()
```

```python
#5. Creating a List of News Articles
news = []
for row in range(0,len(df.index)):
    news.append(' '.join(str(x) for x in df.iloc[row, 2:27]))
```

```python
#6. Verifying Data Type and Content
type(news)
news[0]
```

```python
#7. Text Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(lowercase=True, ngram_range=(1,1))
X = cv.fit_transform(news)
X.shape
```

```python
#8. Preparing Labels
Y = df["News 1"]
Y.shape
```

```python
#9. Splitting Data into Training and Testing Sets
from sklearn.model_selection import train_test_split
print(Y.value_counts())
x_train,x_test,y_train,y_test = train_test_split(X, Y, test_size = 0.3, random_state = 4)
```

```python
#10. Training the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
#11. Predicting and Evaluating the Model
y_pred = model.predict(x_test)
#calculating accuracy
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Select the second news article from the list
news_2 = [news[2]]
# Vectorize the news article using the same CountVectorizer
X_news_2 = cv.transform(news_2)
# Predict the sentiment using the trained model
prediction = model.predict(X_news_2)
# Output the predicted sentiment label
print(prediction[0])
```
